# King County Housing Renovation Analysis

## Business Problem

The median real estate agency has a number of significant business problems. One is how to advise clients hoping to sell their homes on what they can do to maximize the price of their sale. Information on the likely payoff of renovating their homes could be useful to many clients in making a decision as to whether a potential renovtion is worth the cost and time investment.

To aid in making these recommendations we will attempt to answer three questions:

* How much would renovating to improve the condition or grade of the home likely increase the sale price?
* How much would adding an extension to the living area of the home likely increase sale price?
* How much would adding an additional bathroom likely increase sale price?

Constructing a multiple regression model could help to predict the results of any changes that the client chooses to make and the coefficients of the model could allow for rough inferences about the general size of the effect of various changes. This will provide some insight into the answers to these questions.

## Cleaning the Data

Since we are not doing time analysis we will delete the date column. The id number cannot be used in regression and will be removed. We will use zipcodes for location information and so will remove lat/long info as well.

```python
irrelevant = ['id', 'date', 'lat', 'long']
relevant = data.drop(columns=irrelevant)
```

```python
relevant.head()
```

### Dealing with Missing Values

#### *Dealing with Renovation Year Data*

We can see there are 3842 missing values about what year renovations we done, and a large number of zero values as well.

```python
relevant['yr_renovated'].isna().sum()
```

```python
relevant['yr_renovated'].value_counts()
```

We will add a column indicating which values are missing and assume that NaN values are 0 values and represent no renovation. We will also bin the renovations into three categories: "not_renovated", "new_renovations", "old_renovations". We will define new renovations as ones that have been done in the past 10 years. We will then one-hot encode it so it can be used in a regression.

```python
relevant['yr_ren_miss'] = dp.missing_indicator(relevant, 'yr_renovated')
relevant['yr_renovated'] = dp.impute_values(relevant, 'yr_renovated', type='median')
relevant['yr_renovated'].isna().sum()
```

```python
def find_old_new(x, threshold=10):
    if x > (2020-threshold):
        return "new_ren"
    elif x == 0:
        return "not_ren"
    else:
        return "old_ren"
```

```python
relevant['yr_renovated'] = relevant['yr_renovated'].apply(find_old_new)
cats_yr_ren, encode_yr_ren = dp.one_hot_encode(relevant, 'yr_renovated')
relevant[cats_yr_ren] = encode_yr_ren 
```

```python
#One column dropped to avoid dummy variable trap
clean = relevant.drop(columns=['yr_renovated', 'not_ren'])
clean.columns
```

#### *Dealing with Waterfront Data*

There are 2376 null values in the waterfront data and most of the values are 'NO'.

```python
clean['waterfront'].isna().sum()
```

```python
clean['waterfront'].value_counts()
```

For the waterfront data, we will also add a missing column and will assume that the missing values are properties that lack a waterfront. We will also binary encode this column.

```python
clean['waterfront_miss'] = dp.missing_indicator(clean, 'waterfront')
clean = clean.replace({'NO': 0, 'YES': 1})
clean['waterfront'] = dp.impute_values(clean, 'waterfront', type='median')
clean.columns
```

```python
clean['waterfront'].isna().sum()
```

#### *Dealing with View Data*

```python
#One column dropped to avoid dummy variable trap
clean = clean.drop(columns=['sqft_basement', 'no_basement'])
clean.columns
```

## Interpreting the Final Model

#### *Interpreting the Coefficients*

So we have our final model, unfortunately since we log normalized the price and some other data, the coeffients are not as directly interpretable since they are not in the units or scale of the original values.

```python
coef = pd.DataFrame(np.zeros((1, len(X_train_final.columns))))
coef.columns = X_train_final.columns
coef[X_train_final.columns] = final_model.coef_
coef.transpose().head()
```

To make the coefficients interpretable we will transform them.

```python
unnormalized_var = set(coef.columns) - set(cont_columns)
normalized_var = set(cont_columns) - {'price'}
```

```python
for column in unnormalized_var:
    coef[column] = (coef[column].map(np.expm1))*100
```

For all the continuous variables that we normalized, these coefficients represent percent increases in price per 1% increase in the dependent variable. So, for example, a 1% increase in the square footage of the living space is predicted to increase the home's price by about 0.44%

```python
float(coef['sqft_living'])
```

#### Visualizing Square Footage Effects

```python
large = 24
med = 20
small = 12
params = {'axes.titlesize': large,
          'legend.fontsize': med,
          'axes.labelsize': med,
          'xtick.labelsize': med,
          'ytick.labelsize': med,
          'figure.titlesize': large}
plt.rcParams.update(params)
```

```python
x = np.linspace(float(X_train_final[:1]['sqft_living'])-1, float(X_train_final[:1]['sqft_living'])+1)
y = np.expm1(final_model.predict(X_train_final[:1])
             - float(coef['sqft_living'])*float(X_train_final[:1]['sqft_living'])
             + float(coef['sqft_living'])*x)
x_exp = np.expm1(x)
fig, ax = plt.subplots(figsize=(15,8))
ax.set_xlabel("Square Footage of Living Space")
ax.set_ylabel("Predicted Price")
ax.set_title("Effect of Square Footage on Price of Example Home")
ax.scatter(np.expm1(float(X_train_final[:1]['sqft_living'])), np.expm1(final_model.predict(X_train_final[:1])))
plt.plot(x_exp,y)
plt.savefig('images/sqft.png');
```

#### *Evaluating the Error*

The final model has an R-squared of 0.849. This means that the final model explains 85% of the variance in the target variable. to make this more concrete we can find the mean squared error of the model.

```python
X = encoded_df.drop(columns='price')
y = encoded_df['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=0)
train_error, test_error = dp.mean_sq_error(final_model, X_train, y_train, X_test, y_test)
```

These mean square error values are, like the price variable, on a logarithmic scale, so it is harder to get an intuitive feel for the size of the errors in real terms. To address that, let's compare the models predictions on a set of test data, versus its predictions.

```python
preds = np.expm1(final_model.predict(X_test))/1000
unnorm_price = np.expm1(y_test)/1000
fig, ax = plt.subplots(figsize=(15,8))
perfect_line = np.arange(unnorm_price.min(), unnorm_price.max())
ax.plot(perfect_line, perfect_line, linestyle="--", color="orange", label="Perfect Fit")
ax.scatter(unnorm_price, preds, alpha=0.5)
ax.set_xlabel("Actual Value in Thousands of Dollars")
ax.set_ylabel("Predicted Value in Thousands of Dollars")
ax.legend()
plt.savefig('images/prediction_graph.png')
plt.show()
```

## Limitations and Future Work

GIven the collinearity in the model, the coefficients themselves are probably not perfectly reliable when used in inferences, but the predictions of the model if supplied with all the variables will likely make good predictions about the actual home prices.

Technically all this analysis can tell us is the association between renovations and certain house features and prices, it can't actually tell us whether renovating a particular home would lead to a return. With a larger dataset covering more years it might be possible to do an analysis comparing the prices of homes before and after renovations.

The recomendations are also all conditional on the price of making various home renovations. WIth data on what contractors charge for various renovations, it would be possible to produce more solid recommendations from the model.
